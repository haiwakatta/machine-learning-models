```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#read csv file into a dataframe
df = pd.read_csv('Classified Data', index_col=0)
```

```python
#check content of dataframe
df.head()
```

```python
#Need to scale all variables except the target, to reduce scale bias 
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('TARGET CLASS',axis=1))
```

```python
scaler_features = scaler.transform(df.drop('TARGET CLASS',axis=1))
```

```python
#check the scaled features
```

```python
scaler_features
```

```python
#create a new dataframe with the scaled features and column names
df_feat = pd.DataFrame(scaler_features, columns=df.columns[:-1])
```

```python
df_feat
```

```python
#divide dataset into test/train

from sklearn.cross_validation import train_test_split
```

```python
X = df_feat
y = df['TARGET CLASS']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
#initialize KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
#train the model with the training set
knn.fit(X_train, y_train)
```

```python
#predict the test set with the trained model
pred = knn.predict(X_test)
```

```python
pred
```

```python
#Get a report for accuracy of the model
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

```python
#Loop different K values to check with one has the best accuracy
error_rate = []

for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
    
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate, color='blue', linestyle='dashed')
```

```python
error_rate
```

```python
#train the model with the same dataset but k = lowest error rate
knn = KNeighborsClassifier(n_neighbors=36)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
print(confusion_matrix(y_test, pred))

print(classification_report(y_test, pred))
```
