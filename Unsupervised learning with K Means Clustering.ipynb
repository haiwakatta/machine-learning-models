```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#create random data with sklearn
from sklearn.datasets import make_blobs
```

```python
#create 200 data points, with two columns, seperated into 4 groups. 
data = make_blobs(n_samples=200, n_features=2, centers = 4, cluster_std=1.8, random_state=101)
```

```python
#visualize the random data. Color based on group. This is what a 100% KMC model should predict. 
plt.scatter(data[0][:,0],data[0][:,1], c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
#instantiate a KMeans model setting clusters 4. 
kmeans = KMeans(n_clusters = 4)
```

```python
kmeans.fit(data[0])
```

```python
#Get the centers of each cluster
kmeans.cluster_centers_
```

```python
#find the predicted cluster for each row
kmeans.labels_
```

```python
#visualize the predicted model next to the true model
fig , (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1], c=data[1], cmap='rainbow')
```
